```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
import scipy
```

```python
def ploting(Tmax,p,q,t,G):
    y= np.linspace(Tmax,0)
    x=0*y
    plt.plot(x,y)
    x1=np.linspace(0,p)
    b=(q - G*(p**2) -t)/p
    y1=G * x1**2 + b*x1 + t
    plt.plot(x1,y1)
```

```python
def findBest(Tmax,G,p,q):
    t=0
    bestLength= 1000000000000.0
    bestT = t
    while t >= Tmax:
        b = (q - G * (p**2) - t)/p
        if p>0:
            x =np.linspace(0, p, num=50)
        elif p < 0:
            x =np.linspace(p, 0, num=50)
        lot = np.sqrt(1+(2*G*x + ((q-G*(p**2)-t)/p))**2)
        length = scipy.integrate.trapezoid(lot,x)
        if length < bestLength:
            #ploting(Tmax,p,q,bestT,G) #comparing possible lines
            bestLength = length
            bestT = t
        t = t - 0.1 #last 
    ploting(Tmax,p,q,bestT,G)
    print("length: " + str(bestLength) +" t= " + str(bestT))
```

```python
findBest(-100.0,-.9,10.0,-50)
findBest(-100,0.5,-5,-70)
findBest(-100,0,-7,-20)
```
